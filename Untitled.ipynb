```python
import requests
from bokeh.plotting import figure
from bokeh.embed import components
import simplejson as json
import pandas as pd
import numpy as np
from bokeh.io import output_file, output_notebook, show
import matplotlib.pyplot as plt
```

```python
stock = 'GOOG'
selection = ['Close','Open']
```

```python
# Load data:
api_url = 'https://www.quandl.com/api/v1/datasets/WIKI/%s.json?api_key=N_mE5xwVmBJ4hfvDMbsP' % stock
session = requests.Session()
session.mount('http://', requests.adapters.HTTPAdapter(max_retries=3))
raw_data = session.get(api_url)
```

```python
# Clean data:
json_string = json.dumps(raw_data.json(), sort_keys=True)
obj = json.loads(json_string)
```

```python
df_ = pd.DataFrame(np.array(obj['data'],dtype=float),columns=obj['column_names'])
```

```python
pd.to_datetime(list(np.array(obj['data'])[:,0])) #,format="%Y-%m-%d")
index_dates = pd.to_datetime(list(np.array(obj['data'])[:,0]),format="%Y-%m-%d")
df = pd.DataFrame(np.array(obj['data'])[:,1:],dtype=float,columns=obj['column_names'][1:],index=index_dates)
```

```python
df_['Date'] = pd.to_datetime(df_['Date'])
```

```python
from bokeh.models.ranges import Range1d
```

```python
df_['Date']

plot = figure(#tools=TOOLS,
              title='Data from Quandle WIKI set',
              x_axis_label='date',
              x_axis_type='datetime',
              y_axis_label='stock value')

colors = ["red","blue","black","magenta"]
i=0
for sel in selection:
    plot.line(df_['Date'],df_[sel],color=colors[i])
    i+=1

plot.y_range = Range1d(400, 1200)

show(plot)
```
